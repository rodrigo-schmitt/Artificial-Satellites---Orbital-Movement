# Legendre Polynomials - Visualization

This notebook is destined to calculate the Legendre Polynomials using recurrence functions for satellite orbit prediction.

- $P_{n,n} = (2n-1)\cos(\phi)P_{n-1,n-1} $, if $ m = n $(setorial)

- $P_{n,m} = (2n-1)\sin(\phi)P_{n-1,n-1} $, if $m = n - 1 $ (tesseral)

- $P_{n,m} = \frac{(2n-1)}{n-m}sin(\phi)P_{n-1,m} - \frac{(n+m-1)}{n-m}P_{n-2,m} $, if $m \le n$ (tesseral)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def legendre(n, m):
    '''
    Computes the Legendre polynomials using recurrence functions.
    ---
    Arguments:
        n, m - Order of P_nm
    ---
    Returns:
        phi - Angle
        P - P(sin(phi))
    '''
    phi = np.arange(-1, 1, 0.01)
    if m==0 and n==0:
        P = 1
    elif m==0 and n==1:
        P = np.sin(phi)
    elif m==n:
        P = (2*n-1)*np.cos(phi)*legendre(n-1,n-1)[1]
    elif m==n-1:
        P = (2*n-1)*np.sin(phi)*legendre(n-1,n-1)[1]
    elif m<=n:
        P = (2*n-1)/(n-m)*np.sin(phi)*legendre(n-1,m)[1] - \
        (n+m-1)/(n-m)*legendre(n-2,m)[1]
    return phi, P
```

```python
phi, P_33 = legendre(3, 3)
phi, P_66 = legendre(6, 6)
phi, P_84 = legendre(8, 4)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(12,12))
ax1.plot(phi, P_33, label="$P_{3,3}$")
ax2.plot(phi, P_66, 'r', label="$P_{6,6}$")
ax2.plot(phi, P_84, 'g', label="$P_{8,4}$")
fig.suptitle("Legendre Polynomials as functions of $\phi$")
ax1.set_ylabel("P")
ax1.set_xlabel("$\phi$")
ax2.set_ylabel("P")
ax2.set_xlabel("$\phi$")
fig.legend()
plt.show()
```
