# Atmospheric density x Height

Plot the logarithm of atmospheric density as a function of altitude ($log(\rho)\; x \;h$), for 90 < h < 1200 km, and also as a function of the hour angle with the Sun, H (0 < H < 24h).

Consider H = 0, 6, 12 and 18h in the first graph and h = 200,
400, 700 and 1200 km in the second.

Consider the following values for both: latitude f = 0, solar flux F10.7 = 200 and geomagnetic activity Ap = 0.

The simulations were made in NASA's [MSIS-E-90](https://ccmc.gsfc.nasa.gov/modelweb/models/msis_vitmo.php).

```
Input parameters
year= 2000, month= 1, day= 1, hour=(0, 6, 12, 18),
Time_type = Local
Coordinate_type = Geographic
latitude= 0.00, longitude= 0.00, height= 100.00
Prof. parameters: start= 90.00 stop= 1000.00 step= 20.00
Optional parametes: F10.7(daily) =200.; F10.7(3-month avg) =200.; ap(daily) = 0.

```

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
````

````python
import warnings
warnings.filterwarnings("ignore")
````

````python
df = pd.read_excel('atmosphere_data.xlsx')
````

````python
df.drop('Unnamed: 3', axis=1, inplace=True)
````

## Constant solar hour angle, variable height


````python
plt.figure(figsize=(12, 8))

plt.plot(df.iloc[:46]['h (km)'], np.log10(df.iloc[:46]['rho (g/cm-3)']), label=r"$H = 0 h$")
plt.plot(df.iloc[46:46*2]['h (km)'], np.log10(df.iloc[46:46*2]['rho (g/cm-3)']), label=r"$H = 6 h$")
plt.plot(df.iloc[46*2:46*3]['h (km)'], np.log10(df.iloc[46*2:46*3]['rho (g/cm-3)']), label=r"$H = 12 h$")
plt.plot(df.iloc[46*3:]['h (km)'], np.log10(df.iloc[46*3:]['rho (g/cm-3)']), label=r"$H = 18 h$")
plt.xlabel(r"$h (km)$")
plt.ylabel(r"$Log(\rho) \; (g/cm³)$")
plt.title(r"$Log(\rho) \;vs\; h$")
plt.legend()
plt.show()
````

## Constant height, variable solar hour angle

```
Input parameters
year= 2000, month= 1, day= 1, hour= 0.00,
Time_type = Local
Coordinate_type = Geographic
latitude= 0.00, longitude= 0.00, height= (200, 400, 700, 1000)
Prof. parameters: start= 1.00 stop= 24.00 step= 0.50
Optional parametes: F10.7(daily) =200.; F10.7(3-month avg) =200.; ap(daily) = 0.
```

````python
plt.figure(figsize=(12, 8))

plt.plot(df.iloc[:47]['Solar Hour Angle (h).1'], np.log10(df.iloc[:47]['rho (g/cm-3).1']), label=r"$h = 200\;km$")
plt.plot(df.iloc[47:47*2]['Solar Hour Angle (h).1'], np.log10(df.iloc[47:47*2]['rho (g/cm-3).1']), label=r"$h = 400\;km$")
plt.plot(df.iloc[47*2:47*3]['Solar Hour Angle (h).1'], np.log10(df.iloc[47*2:47*3]['rho (g/cm-3).1']), label=r"$h = 700\;km$")
plt.plot(df.iloc[47*3:]['Solar Hour Angle (h).1'], np.log10(df.iloc[47*3:]['rho (g/cm-3).1']), label=r"$h = 1000\;km$")
plt.xlabel(r"$H (h)$")
plt.ylabel(r"$Log(\rho) \; (g/cm³)$")
plt.title(r"$Log(\rho) \;vs\; H$")
plt.legend()
plt.show()
````
